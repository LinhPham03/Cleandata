```python
import pandas as pd
df = pd.read_csv('D:\\raw\\house_price_Dống-Da_Hà-Nội_subdata.csv')
df
```

```python
df.info()
```

```python
df.isnull()
```

```python
df = df.dropna(subset = 'price')
df
```

```python
df = df.fillna(value = {'land_certificate':'Không có thông tin'})
df
```

```python
import numpy as np

from sklearn.impute import SimpleImputer
imputa = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')

imputa.fit(df[['house_direction', 'balcony_direction', 'toilet', 'bedroom', 'floor' ]])
df[['house_direction', 'balcony_direction', 'toilet', 'bedroom', 'floor' ]]= imputa.transform(df[['house_direction', 'balcony_direction', 'toilet', 'bedroom', 'floor' ]])
```

```python
df
```

```python
nha_ngo = df[df['property_type'] == 'trong ngo']
nha_ngo
```

```python
nha_ngo = nha_ngo .dropna(subset = 'area')
nha_ngo 
```

```python
nha_ngo['price/m2'] = nha_ngo['price']/ nha_ngo['area']
nha_ngo
```

- IQR

```python
Q1 = nha_ngo.quantile(0.25)
Q3 = nha_ngo.quantile(0.75)
IQR = Q3-Q1

Q1 - 1.5 * IQR
```

```python
outliner = ((nha_ngo < (Q1 - 1.5 * IQR)) | (nha_ngo > (Q3+ 1.5* IQR)))
outliner
```

```python
nha_ngo = nha_ngo[~ ((nha_ngo < (Q1 - 1.5 * IQR)) | (nha_ngo > (Q3+ 1.5* IQR))).any(axis = 1)]
print(nha_ngo)
```
